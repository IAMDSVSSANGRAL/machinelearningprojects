## Popularity based reccomendation system

```python
ratings_with_name = ratings.merge(books, on = 'ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating':'num_rating'}, inplace = True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns = {'Book-Rating':'Avg_rating'}, inplace = True)
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
popularity_df
```

```python
popular_df = popularity_df[popularity_df['num_rating']>=250].sort_values('Avg_rating',ascending = False).head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books,on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','num_rating','Avg_rating','Image-URL-M']]
```

## Colabrative filtering

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
active_user = x[x].index
```

```python
filtered_Rating = ratings_with_name[ratings_with_name['User-ID'].isin(active_user)]
filtered_Rating.head()
```

```python
y = filtered_Rating.groupby('Book-Title').count()['Book-Rating']>50
famous_books = y[y].index
famous_books
```

```python
final_Ratings = filtered_Rating[filtered_Rating['Book-Title'].isin(famous_books)]
final_Ratings
```

```python
pt = final_Ratings.pivot_table(index = 'Book-Title',columns = 'User-ID', values= 'Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace= True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_item = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:11]
    
    
    for i in similar_item:
        print(pt.index[i[0]])
```

```python
recommend('A Bend in the Road')
```

## recommendation function for ui

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_item = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_item:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('A Bend in the Road')
```
